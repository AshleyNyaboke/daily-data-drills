```python
import numpy as np

# input
input_value = np.array([[0,0],[0,1],[1,1],[1,0]])

# output
output = np.array([[0],[1],[1],[0]])

# weights and bias
weights = np.array([[0.1], [0.2]])
bias = 0.3


def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# training
for epoch in range(10000):

    # forward pass
    weighted_sum = np.dot(input_value, weights) + bias
    y_pred = sigmoid(weighted_sum)

    # error
    error = y_pred - output

    # derivative of sigmoid
    d_sigmoid = y_pred * (1 - y_pred)

    # backprop
    derivative = error * d_sigmoid

    # update weights
    weights -= 0.05 * np.dot(input_value.T, derivative)

    # update bias
    bias -= 0.05 * np.sum(derivative)

print(weights)
print(bias)
```

```python
# prediction
pred = np.array([0,1])
result = sigmoid(np.dot(pred, weights) + bias)
print(result)
```
